**Question 1**
In a simple linear regression model, we aim to predict an outcome variable Y using a predictor variable X. The model consists of a slope β1, which represents the rate of change in Y for each unit increase in X , and an intercept β0, which is the value of Y when X is zero. Additionally, an error term ϵ is included to account for the variability in Y that is not explained by X. This error term is assumed to follow a normal distribution with a mean of zero and a variance σ^2, making Y follow a normal distribution centered at β0 + β1 X with variance σ^2 for any given X. The regression model formula is Y = β0+ β1X +ϵ.

```python
#Question 1
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Set regression coefficients and intercept
beta_0 = 2.0  # Intercept
beta_1 = 1.5  # Slope
sigma = 1.0   # Standard deviation of the error term

# Generate sample data for the predictor variable X
np.random.seed(0)
X = np.linspace(0, 10, 50)

# Generate values for the outcome variable Y based on the regression model
epsilon = norm.rvs(0, sigma, size=len(X))  # Error term, following a normal distribution
Y = beta_0 + beta_1 * X + epsilon          # Generate outcome variable Y

# Plot a scatter plot of X and Y
plt.scatter(X, Y, label='Sample Data')
plt.plot(X, beta_0 + beta_1 * X, color='red', label='True Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression Model')
plt.legend()
plt.show()
```

**Question 2**
We generated a simulated data set containing a linear relationship, assuming that the model is 𝑌=3+2𝑋+𝜖 (𝜖 is a normally distributed random noise). Next, we used the ols function in statsmodels.formula.api to fit a simple linear regression model and output a statistical summary of the regression results. Finally, we used matplotlib to visualize the data points and the fitted regression line to intuitively demonstrate the regression effect.

```python
#Question 2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Step 1: Simulate the dataset
np.random.seed(0)
X = np.random.rand(50) * 10  # Random values for X between 0 and 10
epsilon = np.random.normal(0, 1, 50)  # Normally distributed noise
Y = 3 + 2 * X + epsilon  # Linear relation with noise

# Create a DataFrame
data = pd.DataFrame({'X': X, 'Y': Y})

# Step 2: Fit the Simple Linear Regression Model
model = smf.ols('Y ~ X', data=data).fit()
print(model.summary())

# Step 3: Visualize the data and fitted model
plt.figure(figsize=(10, 6))
plt.scatter(data['X'], data['Y'], color='blue', label='Data Points')
plt.plot(data['X'], model.predict(data['X']), color='red', label='Fitted Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression: Fitted Line')
plt.legend()
plt.show()
```

In the ordinary least squares (OLS) method, the fitted line is chosen to minimize the sum of the squared differences between observed data points and the predicted values on the line. These squared differences (or "residuals") are used to avoid canceling out positive and negative errors and to emphasize larger errors, thus ensuring the best possible fit for the data by reducing overall prediction error.

```python
#Question 9
import pandas as pd
import statsmodels.api as sm

# Define the short wait limits and prepare to collect results
short_wait_limits = [62, 64, 66]
results = {}

# Perform analysis for each short wait limit
for limit in short_wait_limits:
    # Filter the data based on the current limit
    short_wait_data = data[data['X'] < limit]  # 'X' is used instead of 'wait_time'
    
    # Ensure there's data after filtering
    if short_wait_data.empty:
        print(f"No data available for wait times less than {limit} minutes.")
        results[limit] = None
        continue
    
    # Set up the variables for regression
    X = short_wait_data['X']  # 'X' instead of 'wait_time'
    y = short_wait_data['Y']  # 'Y' instead of 'duration'
    X = sm.add_constant(X)  # Adds a constant term to the predictor
    
    # Fit the linear regression model
    model = sm.OLS(y, X).fit()
    
    # Extract and store the p-value for the slope
    p_value = model.pvalues['X']
    results[limit] = p_value

# Organize the results into a DataFrame and interpret the evidence
results_df = pd.DataFrame(list(results.items()), columns=['Short Wait Limit', 'P-Value'])
results_df['Evidence Against Null Hypothesis'] = results_df['P-Value'].apply(
    lambda p: 'Strong Evidence' if p is not None and p < 0.05 else 'Weak or No Evidence'
)

# Display the results in the output
print("Evidence Against Null Hypothesis")
print(results_df)
```

```python
#Question 10
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

# Define your threshold for long wait times
threshold = 70  # Adjust based on your criteria
long_wait = old_faithful['waiting'] > threshold

# Step 1: Bootstrap Sampling Distribution of Slope Coefficients
n_bootstrap_samples = 1000
bootstrap_slopes = []

for _ in range(n_bootstrap_samples):
    bootstrap_sample = old_faithful[long_wait].sample(n=160, replace=True)
    model = smf.ols('eruptions ~ waiting', data=bootstrap_sample).fit()
    bootstrap_slopes.append(model.params['waiting'])

# Plotting the bootstrap slope distribution
plt.hist(bootstrap_slopes, bins=30, edgecolor='k')
plt.title("Bootstrap Sampling Distribution of Slope Coefficients")
plt.xlabel("Slope Coefficient")
plt.ylabel("Frequency")
plt.show()

# Step 2: Simulate Samples Under Null Hypothesis
np.random.seed(0)
n_simulations = 1000
null_slope_coefficients = []

for _ in range(n_simulations):
    x_vals = old_faithful[long_wait]['waiting']
    y_simulated = 1.65 + 0 * x_vals + np.random.normal(0, 0.37, size=len(x_vals))
    sim_data = pd.DataFrame({'waiting': x_vals, 'eruptions': y_simulated})
    model_sim = smf.ols('eruptions ~ waiting', data=sim_data).fit()
    null_slope_coefficients.append(model_sim.params['waiting'])

# Plotting the null distribution of slope coefficients
plt.hist(null_slope_coefficients, bins=30, edgecolor='k')
plt.title("Null Distribution of Slope Coefficients")
plt.xlabel("Slope Coefficient")
plt.ylabel("Frequency")
plt.show()

# Step 3: Calculate 95% CI and Compare p-values
lower_bound, upper_bound = np.percentile(bootstrap_slopes, [2.5, 97.5])
print("95% CI for bootstrapped slope coefficients:", (lower_bound, upper_bound))

contains_zero = lower_bound <= 0 <= upper_bound
print("Does the 95% CI contain 0?", contains_zero)

original_model = smf.ols('eruptions ~ waiting', data=old_faithful[long_wait]).fit()
original_p_value = original_model.pvalues['waiting']

null_p_value = np.mean([abs(slope) >= abs(original_model.params['waiting']) for slope in null_slope_coefficients])

print("Original model p-value:", original_p_value)
print("Simulated null p-value:", null_p_value)
```
